```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
# generate synthetic two-dimensional data
X, y = make_blobs(random_state=1)
# build the clustering model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
print("Cluster memberships:\n{}".format(kmeans.labels_))
```

```python
print(kmeans.predict(X))
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], color='r', s=100)
# Title of the plot
plt.title('KMeans')
# Remove x-axis ticks
plt.xticks(())
# Remove y-axis ticks
plt.yticks(())
# Show the plot
```

```python
from sklearn.cluster import AgglomerativeClustering
 
X, y = make_blobs(random_state=1)
agg = AgglomerativeClustering(n_clusters=3)
assignment = agg.fit_predict(X)
agg.labels_
```

```python
from matplotlib import pyplot as plt 
plt.scatter(X[:, 0], X[:, 1], c=assignment)
# Title of the plot
plt.title('Agglomerative Clustering')
plt.xlabel('Figure 1')
plt.ylabel('Figure 2')
```

```python
from sklearn.datasets import make_blobs
# Import the dendrogram function and the ward clustering function 
from scipy.cluster.hierarchy import dendrogram, ward
X, y = make_blobs(random_state=0, n_samples=12)
# Apply the ward clustering to the data array X
# The SciPy ward function returns an array that specifies the distances
# bridged when performing agglomerative clustering
linkage_array = ward(X)
```

```python
dendrogram(linkage_array)
```

```python
from matplotlib import pyplot as plt
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [7.25, 7.25], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')
ax.text(bounds[1], 7.25, ' two clusters', va='center', fontdict={'size': 
15})
ax.text(bounds[1], 4, ' three clusters', va='center', fontdict={'size': 
15})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```
